```python
from IPython.display import display

data = cdata
km = pd.read_csv(f"./results/km_{data['name']}.csv")
em = pd.read_csv(f"./results/em_{data['name']}.csv")
method = "PCA"
km_pca = pd.read_csv(f"./results/km_{method.lower()}_{data['name']}.csv")
em_pca = pd.read_csv(f"./results/em_{method.lower()}_{data['name']}.csv")
rerror_pca = pd.read_csv(f"./results/rerror_{method.lower()}_{data['name']}.csv")
method = "ICA"
km_ica = pd.read_csv(f"./results/km_{method.lower()}_{data['name']}.csv")
em_ica = pd.read_csv(f"./results/em_{method.lower()}_{data['name']}.csv")
rerror_ica = pd.read_csv(f"./results/rerror_{method.lower()}_{data['name']}.csv")
method = "RP"
km_rp = pd.read_csv(f"./results/km_{method.lower()}_{data['name']}.csv")
em_rp = pd.read_csv(f"./results/em_{method.lower()}_{data['name']}.csv")
rerror_rp = pd.read_csv(f"./results/rerror_{method.lower()}_{data['name']}.csv")
method = "FA"
km_fa = pd.read_csv(f"./results/km_{method.lower()}_{data['name']}.csv")
em_fa = pd.read_csv(f"./results/em_{method.lower()}_{data['name']}.csv")
rerror_fa = pd.read_csv(f"./results/rerror_{method.lower()}_{data['name']}.csv")
                        
n_clusters = list(range(2, data['n_features']+1))
                        
cluster = 2
km_perf = km[km['n_clusters']==ncluster]
pca_perf = km_pca[km_pca['n_clusters']==ncluster]
ica_perf = km_ica[km_ica['n_clusters']==ncluster]
rp_perf = km_rp[km_rp['n_clusters']==ncluster]                        
fa_perf = km_fa[km_fa['n_clusters']==ncluster]    

vals = []
metrics = ['AMI','silhouette','time']
for col in metrics:
    vals.append({
        '_K-Means': km_perf[col].values[0],
        'PCA': pca_perf[col].values[0],
        'ICA': ica_perf[col].values[0],
        'RP': rp_perf[col].values[0],
        'FA': fa_perf[col].values[0]
    })
df = pd.DataFrame(vals, index=metrics)
display(df)

em_perf = em[em['n_clusters']==ncluster]
pca_perf = em_pca[em_pca['n_clusters']==ncluster]
ica_perf = em_ica[em_ica['n_clusters']==ncluster]
rp_perf = em_rp[em_rp['n_clusters']==ncluster]                        
fa_perf = em_fa[em_fa['n_clusters']==ncluster]    

vals = []
metrics = ['AMI','silhouette','time']
for col in metrics:
    vals.append({
        '_EM': em_perf[col].values[0],
        'PCA': pca_perf[col].values[0],
        'ICA': ica_perf[col].values[0],
        'RP': rp_perf[col].values[0],
        'FA': fa_perf[col].values[0]
    })
df = pd.DataFrame(vals, index=metrics)
display(df)
```
